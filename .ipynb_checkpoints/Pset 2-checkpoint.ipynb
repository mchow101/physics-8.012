```python
#imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
#spring force function
def force(x, k=0.1, b=0, v=0):
    return -k * x - b * v
```

```python
#plot velocity and position
def plot_damped(k, b, title):
    #variables
    t_arr = []
    x_arr = []
    v_arr = []
    dt = 0.01
    m = 1.0
    x_old = 1
    v_old = -1
    #numerically compute the position and velocity
    for i in range(5500):
        t = i * dt
        a = force(x_old, k, b, v_old) / m
        v = v_old + dt * a
        x = x_old + dt * v_old + 0.5 * dt * dt * a
        t_arr.append(t)
        x_arr.append(x)
        v_arr.append(v)
        x_old, v_old = x, v 
    #plot the figures
    fig, ax = plt.subplots()
    ax.plot(t_arr, x_arr, label='x(t)')
    ax.plot(t_arr, v_arr, label='v(t)')
    ax.set(xlabel='time in seconds', ylabel='x(t), v(t)', title=title)
    ax.grid()
    plt.legend()
    plt.show()
```

```python
#problem 8 plots
plot_damped(1, 0, "Spring Only")
plot_damped(1, 0.2, "Damped Spring")
plot_damped(0, 0.2, "Damped Only")
```
